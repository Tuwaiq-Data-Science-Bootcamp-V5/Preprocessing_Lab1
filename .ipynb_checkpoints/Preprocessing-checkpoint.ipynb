# Explore the distribution of the target variable

```python
df['term_deposit'] = df['term_deposit'].astype(int)
df['term_deposit']
```

```python
sns.histplot(df['term_deposit'],bins=3)
```

# Apply oversampling or SMOTE to the target label

```python
X = df.drop(columns = 'term_deposit')
X = X.drop(columns = 'user_id')
y = df['term_deposit']

smote = SMOTE()
newx, newy = smote.fit_resample(X, y)
dff = pd.concat([pd.DataFrame(newx), pd.Series(newy)], axis=1)
dff
```

```python
sns.histplot(dff['term_deposit'],bins=3)
```

# Normalize the data

```python
scaler = MinMaxScaler()
X_resampled_normalized = scaler.fit_transform(dff)
X_resampled_normalized = pd.DataFrame(X_resampled_normalized,columns=dff.columns)
X_resampled_normalized
```

# Standardlize the data

```python
scaler = StandardScaler()
scaled_df = scaler.fit_transform(dff)
scaled_df = pd.DataFrame(scaled_df,columns=dff.columns)
scaled_df
```

# Plot the PCA

```python
pca = PCA()
pca.fit(X_resampled_normalized)
plt.figure(figsize=(14,7))
exp_var = pd.Series(pca.explained_variance_ratio_) 
exp_var.plot(kind="bar", alpha=0.7)

total = 0
var_ls = []
for x in exp_var:
    total = total+x
    var_ls.append(total)

pd.Series(var_ls).plot(marker="o", alpha=0.7)
plt.xlabel("Principle Components", fontsize="x-large")
plt.ylabel("Percentage Variance Explained", fontsize="x-large")
plt.title("Diabetes Scree Plot", fontsize="xx-large")
plt.show();
```
