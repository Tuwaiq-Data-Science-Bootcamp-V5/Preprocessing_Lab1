# Apply oversampling or SMOTE to the target label

```python
X = df.drop(columns=['user_id','term_deposit'])
y = df['term_deposit']

smote = SMOTE(random_state=40)
X1, y1 = smote.fit_resample(X, y)

df_new= pd.DataFrame(X1, columns=X.columns)
df_new['term_deposit'] = y1

resampled = df_new['term_deposit'].value_counts() / len(df_new)
print(resampled)
```

# Normalize the data

```python
scaler = MinMaxScaler()
normalized = scaler.fit_transform(X)
df_new2 = pd.DataFrame(normalized, columns=X.columns)

df_new2['term_deposit'] = y
df_new2.head(15)
```

# Standardlize the data

```python
scaler = StandardScaler()
X_stand = scaler.fit_transform(X)
df_new3 = pd.DataFrame(X_stand, columns=X.columns)

df_new3['term_deposit'] = y
df_new3.head(15)
```

# Plot the PCA

```python
X_stand = df_new3.drop(columns=['term_deposit'])
pca = PCA()
pca.fit(X_stand)

plt.figure(figsize=(14,7))
exp_var = pd.Series(pca.explained_variance_ratio_) 
exp_var.plot(kind="bar", alpha=0.7)

total = 0
var_ls = []
for x in exp_var:
    total = total + x
    var_ls.append(total)


pd.Series(var_ls).plot(marker="o", alpha=0.7)
plt.xlabel("Principle Components", fontsize="x-large")
plt.ylabel("Percentage Variance Explained", fontsize="x-large")
plt.title("Diabetes Scree Plot", fontsize="xx-large")
plt.show();
```
