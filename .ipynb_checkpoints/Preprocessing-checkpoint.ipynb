# Apply oversampling or SMOTE to the target label

```python
df.drop(columns=['user_id'],axis=1,inplace=True)
df.columns
```

```python
x=df
y=df['term_deposit']
oversampled = SMOTE(random_state=0)
X_train_smote, y_train_smote = oversampled.fit_resample(x, y)
y_train_smote.value_counts()
```

# Normalize the data

```python
col=['number_transactions', 'total_amount_usd', 'job_management',
       'job_technician', 'job_entrepreneur', 'job_blue-collar', 'job_retired',
       'job_admin.', 'job_services', 'job_self-employed', 'job_unemployed',
       'job_housemaid', 'job_student', 'education_tertiary',
       'education_secondary', 'education_Unknown', 'education_primary',
       'default', 'housing', 'loan', 'contact_cellular', 'contact_telephone',
       'duration', 'campaign', 'pdays', 'previous', 'term_deposit',
       'device_desktop', 'device_tablet', 'single', 'age_group_encoded',
       'month_joined']
```

```python
scaler = MinMaxScaler().fit(X_train_smote)
```

```python
scaled_data = scaler.transform(X_train_smote)

scaled_df = pd.DataFrame(scaled_data,columns=col)
scaled_df
```

```python
scaled_df.describe().round()
```

# Standardlize the data

```python
scaler = StandardScaler().fit(X_train_smote)
```

```python
standarizes_data = scaler.fit_transform(X_train_smote)
standarizes_data = pd.DataFrame(standarizes_data,columns=col)
standarizes_data
```

```python
standarizes_data.describe().round()
```

# Plot the PCA

```python
pca = PCA()
pca.fit(scaled_df)
```

```python
pd.Series(pca.explained_variance_ratio_) 
```

```python
#plt.style.use("seaborn")
plt.figure(figsize=(14,7))
exp_var = pd.Series(pca.explained_variance_ratio_) 
exp_var.plot(kind="bar", alpha=0.7)

total = 0
var_ls = []
for x in exp_var:
    total = total+x
    var_ls.append(total)

pd.Series(var_ls).plot(marker="o", alpha=0.7)
plt.xlabel("Principle Components", fontsize="x-large")
plt.ylabel("Percentage Variance Explained", fontsize="x-large")
plt.title("Diabetes Scree Plot", fontsize="xx-large")
plt.show();
```
