# Apply oversampling or SMOTE to the target label

```python
df.drop(columns="user_id",axis=1,inplace=True)
df
```

```python
xx = df 
yy =df["term_deposit"]
smote = SMOTE(random_state=0)
xx, yy = smote.fit_resample(xx, yy)
yy.value_counts()
```

# Normalize the data

```python
scaler = MinMaxScaler().fit(xx)
scaled_data = scaler.transform(xx)
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)
scaled_df
```

# Standardlize the data

```python
scaler = StandardScaler().fit(xx)
standarizes_data = scaler.fit_transform(xx)
standarizes_data = pd.DataFrame(standarizes_data, columns=df.columns)
standarizes_data
```

# Plot the PCA

```python
pca = PCA()
pca.fit(scaled_df)
```

```python
pd.Series(pca.explained_variance_ratio_) 
```

```python
plt.figure(figsize=(14,7))
exp_var = pd.Series(pca.explained_variance_ratio_) 
exp_var.plot(kind="bar", alpha=0.7)

total = 0
var_ls = []
for x in exp_var:
    total = total+x
    var_ls.append(total)

pd.Series(var_ls).plot(marker="o", alpha=0.7)
plt.xlabel("Principle Components", fontsize="x-large")
plt.ylabel("Percentage Variance Explained", fontsize="x-large")
plt.title("PCA", fontsize="xx-large")
plt.show();
```
